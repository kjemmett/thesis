```python
import numpy as np
from scipy.spatial.distance import pdist, squareform

import matplotlib.pyplot as plt
from matplotlib import collections as mc
from matplotlib.patches import Circle


%matplotlib inline
```

```python
r1 = 96
eps1 = 5
theta1 = np.random.uniform(0, 2*np.pi, 50)

r2 = 24
eps2 = 5
theta2 = np.random.uniform(0, 2*np.pi, 50)

x1 = r1*np.cos(theta1)+eps1*np.random.randn(theta1.shape[0])
y1 = r1*np.sin(theta1)+eps1*np.random.randn(theta1.shape[0])

x2 = r1 + r2 + r2*np.cos(theta2)+eps2*np.random.randn(theta2.shape[0])
y2 = r2*np.sin(theta2)+eps2*np.random.randn(theta2.shape[0])

x = np.hstack([x1,x2])
y = np.hstack([y1,y2])

# distance matrix
d = squareform(pdist(np.array((x,y)).T))

fig, ax = plt.subplots(3, 2, figsize=(10, 13))
sizes = [2, 4, 8, 16, 24, 48]

for n in range(6):
    px = n % 3
    py = n / 3

    ax[px,py].set_aspect('equal')
    
    # add 0-simplices
    for xx, yy in zip(x, y):
        ax[px,py].add_artist(Circle(xy=(xx, yy), radius=sizes[n], linewidth = 0, color=(.39, .77, 1.), alpha=0.25))
    ax[px, py].scatter(x,y, c='k', linewidths = 0, alpha=1, s=2)
    
    # add 1-simplices
    lines = []
    for i in range(x.shape[0]):
        for j in range(i, x.shape[0]):
            if (d[i,j] < 2*sizes[n]):
                lines.append([(x[i],y[i]),(x[j],y[j])])
    
    lc = mc.LineCollection(lines, color='k', linewidths=.1)
    ax[px,py].add_collection(lc)
   
    # add 2-simplices
    triangles = []
    
    mc.PolyCollection(triangles, color = 'g', linewidths=0)
    
    # finalize subfigure
    ax[px,py].axis('off')
    ax[px,py].set_title('scale={0}'.format(sizes[n]))
    
plt.tight_layout()
# plt.savefig('./expanding_balls.pdf', format='pdf')
```
